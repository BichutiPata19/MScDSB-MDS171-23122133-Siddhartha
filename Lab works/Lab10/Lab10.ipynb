```python
class Student:
    def __init__(self):
        self.students = {}

    def add_student(self, roll_number, name):
        if roll_number not in self.students:
            self.students[roll_number] = {"name": name, "marks": {}}
            print(f"Student {name} with Roll Number {roll_number} added.")
        else:
            print(f"Student with Roll Number {roll_number} already exists.")

    def enter_marks(self, roll_number, subject, score):
        if roll_number in self.students:
            self.students[roll_number]["marks"][subject] = score
            print(f"Mark entered for {subject}: {score}")
        else:
            print(f"Student with Roll Number {roll_number} not found.")

    def view_student_details(self, roll_number):
        if roll_number in self.students:
            student = self.students[roll_number]
            print(f"Student Details for Roll Number {roll_number}:")
            print(f"Name: {student['name']}")
            print("Marks:")
            for subject, score in student['marks'].items():
                print(f"{subject}: {score}")
        else:
            print(f"Student with Roll Number {roll_number} not found.")

def main():
    student_manager = Student()

    while True:
        print("\nStudent Management System")
        print("1. Add Student")
        print("2. Enter Marks")
        print("3. View Student Details")
        print("4. Quit")

        choice = input("Enter your choice: ")

        if choice == "1":
            roll_number = input("Enter Roll Number: ")
            name = input("Enter Name: ")
            student_manager.add_student(roll_number, name)
        elif choice == "2":
            roll_number = input("Enter Roll Number: ")
            subject = input("Enter Subject: ")
            score = input("Enter Score: ")
            student_manager.enter_marks(roll_number, subject, score)
        elif choice == "3":
            roll_number = input("Enter Roll Number: ")
            student_manager.view_student_details(roll_number)
        elif choice == "4":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```
